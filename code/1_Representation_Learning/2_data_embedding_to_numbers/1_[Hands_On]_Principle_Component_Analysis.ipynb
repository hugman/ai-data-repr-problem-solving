This notebook demonstrates the application of PCA on a 2D dataset.
PCA is a dimensionality reduction technique that identifies the directions (principal components) along which the data varies the most.

## Steps:
 1. Generate synthetic 2D data from a Gaussian distribution.
 2. Apply PCA to extract the principal components.
 3. Transform the data into the principal component space.
 4. Visualize the original data and overlay the principal components as vectors.
 
 The generated scatter plot will show:
 - Original data points.
 - Principal component vectors originating from the data mean.
 - Reference lines for clarity.
 
 This helps illustrate how PCA identifies the most significant directions of variation.

## Import Necessary Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

## Generate Synthetic Data
Generate a 2D dataset from a multivariate normal distribution.

```python
# Set random seed for reproducibility
np.random.seed(42)
```

```python
# Define mean and covariance matrix for a 2D Gaussian distribution
mean = [2, 2]  # Mean vector (center of distribution)
cov = [[3, 1], [1, 2]]  # Covariance matrix defining the spread of data

# Generate 100 random samples from the multivariate normal distribution
X = np.random.multivariate_normal(mean, cov, 100)
```

## Apply Principal Component Analysis (PCA)
Fit the PCA model and extract the principal components.


```python
# Apply Principal Component Analysis (PCA)
pca = PCA(n_components=2)  # Create PCA model for 2D data (full dimension)
pca.fit(X)  # Compute principal components
X_pca = pca.transform(X)  # Transform data into the principal component space
components = pca.components_  # Extract the principal component vectors
```

## Visualize Original Data with Principal Components
Scatter plot of the original data points and principal component vectors.


```python
# Plot the original data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], alpha=0.6, label="Original Data", edgecolors='k')

# Compute the mean (center of data) to serve as the origin for principal components
origin = np.mean(X, axis=0)

# Draw principal component vectors (scaled for better visualization)
plt.quiver(*origin, *components[0] * 2, color='r', angles='xy', scale_units='xy', scale=1, label="PC1")
plt.quiver(*origin, *components[1] * 2, color='g', angles='xy', scale_units='xy', scale=1, label="PC2")

# Add reference lines for better readability
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)

# Set labels and title
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Principal Component Analysis (PCA)")
plt.legend()
plt.grid(alpha=0.3)
```
